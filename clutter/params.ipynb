```python
from exponential import *
from scipy.optimize import minimize
```

```python
params=np.load("params.npy",allow_pickle=True)
```

```python
def f(x):
        
        if x[0]<=0 or x[1]<=0 or x[2]<=0:
            return 10000
        newparams=np.append(params,np.expand_dims(np.array(x),1)*mp.mpf(1),axis=1)


        subspace=Subspace(newparams.shape[1])

        subspace.set_N_func(N_func)
        subspace.set_H_func(H_func)

        subspace.set_params(newparams)

        subspace.make_N_mat()
        subspace.make_H_mat()
        subspace.find_N_eigens()
        subspace.make_Y_mat()
        subspace.make_invs_sqrt_beta_mats()
        subspace.make_P_mats()
        subspace.find_P_eigens()
        subspace.find_energy_levels()
        return subspace.energy_levels[0]
```

```python
for i in range(100):
    print()
    print(params.shape[1],"+1")
    x0=np.abs(np.array([400,1,0.02])*np.exp(np.random.normal(0,2,3)))
    print(x0)
    res=minimize(f,x0=x0,method='L-BFGS-B')
    params=np.append(params,np.expand_dims(np.array(x0),1)*mp.mpf(1),axis=1)
    print(res)
```

```python
np.save("params",params,allow_pickle=True)
```
