```python
import numpy as np
import mpmath as mp

params=np.load("data/best100params.npy",allow_pickle=True)
```

```python
Norms=np.zeros(100)
for i in range(100):
    Norms[i]=mp.norm(params[:,i])
```

```python
order=np.argsort(Norms)
```

```python
params=params[:,order]
```

```python
for i in range(100):
    print(mp.norm(params[:,i]))
```

```python
np.save("data/best100params.npy",params,allow_pickle=True)
```
