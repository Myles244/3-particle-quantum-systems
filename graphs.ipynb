```python
from exponential import *
import matplotlib.pyplot as plt
```

```python
#produce graph of convergence for each of the sets of paramiters

#load in data
N, E, D =np.array(np.load('data/convergence.npy',allow_pickle=True),dtype=float).transpose()
altN, altE, altD =np.array(np.load('data/altconvergence.npy',allow_pickle=True),dtype=float).transpose()
altaltN, altaltE, altaltD =np.array(np.load('data/altaltconvergence.npy',allow_pickle=True),dtype=float).transpose()

#graph of energy convergence
plt.scatter(N,E,color='C0',label='Energy')
plt.scatter(altN,altE,color='C2',label='Novel Contribution')
plt.scatter(altaltN,altaltE,color='C1',label='The Delta Term')

plt.plot([N[0],N[-1]],[E[-1],E[-1]],linestyle='--',color='C0')
plt.plot([altN[0],altN[-1]],[altE[-1],altE[-1]],linestyle='--',color='C2')
plt.plot([altaltN[0],altaltN[-1]],[altaltE[-1],altaltE[-1]],linestyle='--',color='C1')

plt.ylabel('Ground State Energy')
plt.xlabel('Number of basis states')

plt.ylim([-402.6373022,-402.6373003])
yticks=np.round([-402.6373020,-402.6373015,-402.6373010,-402.6373005],decimals=7)
ylabel=[]
for ytick in yticks:
    ylabel.append(np.format_float_positional(ytick,precision=7,unique=False))
np.set_printoptions(precision=3)
plt.yticks(yticks,ylabel)

plt.legend()
```

```python
#graph of delta term convergence
plt.scatter(N,D,color='C0',label='Energy')
plt.scatter(altN,altD,color='C2',label='Novel Contribution')
plt.scatter(altaltN,altaltD,color='C1',label='The Delta Term')

plt.plot([N[0],N[-1]],[D[-1],D[-1]],linestyle='--',color='C0')
plt.plot([altN[0],altN[-1]],[altD[-1],altD[-1]],linestyle='--',color='C2')
plt.plot([altaltN[0],altaltN[-1]],[altaltD[-1],altaltD[-1]],linestyle='--',color='C1')

plt.ylabel('Expectation of the Delta')
plt.xlabel('Number of basis states')

plt.legend()

plt.ylim([0.31,0.32])
```
