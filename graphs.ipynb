```python
from exponential import *
import matplotlib.pyplot as plt
```

```python
#produce graph of convergence for each of the sets of paramiters

#load in data
N, E, D =np.real(np.array(np.load('data/bestconvergence.npy',allow_pickle=True),dtype=complex)).transpose()

#graph of energy convergence
plt.scatter(N,E,color='C0',label='Energy')

#plt.plot([N[0],N[-1]],[E[-1],E[-1]],linestyle='--',color='C0')

plt.ylabel('Ground State Energy')
plt.xlabel('Number of basis states')

plt.ylim([-402.6373022,-402.6372998])
yticks=np.round([-402.6373020,-402.6373015,-402.6373010,-402.6373005,-402.6373000],decimals=7)
ylabel=[]
for ytick in yticks:
    ylabel.append(np.format_float_positional(ytick,precision=7,unique=False))
np.set_printoptions(precision=3)
plt.yticks(yticks,ylabel)

plt.savefig("../Computing-project-report/Figures/energy-convergence.pdf")
```

```python
#graph of delta term convergence
plt.scatter(N,D,color='C0',label='Energy')
#plt.plot([N[0],N[-1]],[D[-1],D[-1]],linestyle='--',color='C0')

plt.ylabel('Expectation of the Delta')
plt.xlabel('Number of basis states')

plt.savefig("../Computing-project-report/Figures/delta-convergence.pdf")
```
