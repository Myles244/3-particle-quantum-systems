```python
import numpy as np
import matplotlib.pyplot as plt

# Load in data
N, E, D = np.real(np.array(np.load('data/bestconvergence.npy', allow_pickle=True), dtype=complex)).transpose()

# Main graph of energy convergence
fig, ax = plt.subplots()
scatter = ax.scatter(N, E, color='C0')

# Plot literature values
ax.plot([N[0], N[-1]], [-402.637263035135314, -402.637263035135314], color='C1', linestyle='--', label='Pachucki')
ax.plot([N[0], N[-1]], [-402.63726963219007450, -402.63726963219007450], color='C2', linestyle='--', label='Frolov')

# Format x-axis
ax.set_xlabel('Number of basis states')

# Format y-axis
ax.set_ylabel('Ground State Energy (Hartrees)')
yticks = [-402.6373, -402.6372, -402.6371, -402.6370]
ax.set_yticks(yticks)
ax.set_yticklabels(yticks)

ax.legend()

# Add inset zoom plot
x1, x2, y1, y2 = N[7]- np.abs(N[7]-N[100])/20, N[100], E[100]- np.abs(E[7]-E[100])/10, E[7]   # Define zoom region (update values as needed)
axins = ax.inset_axes([0.3, 0.4, 0.67, 0.57])  # Adjust position and size of inset
axins.scatter(N, E, color='C0')  # Scatter in inset

# Set limits and remove tick labels for inset
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.set_xticks([])
axins.set_yticks([])

# Add rectangle and connecting lines between axes
_,conectors=ax.indicate_inset_zoom(axins, edgecolor="black")
conectors[0].set_visible(False)
conectors[1].set_visible(True)
conectors[2].set_visible(True)
conectors[3].set_visible(False)
# Show the plot
plt.show()
```
